## Variables, Operadores y Funciones

### Variables

**Variable**

Las variables son el nombre que asigna a la posición de la memoria en la que se alojan ciertos datos o valores.

**Tipos de datos**

- int = números enteros (1982).
- float = números decimales (0,321)
- str = cadenas de caracteres siempre escritas entre comillas simples o dobles ("Hola", "22", "+"). 
- bool = los booleanos sirven para evaluar si una condición es o no es cumplida y actuar en consecuencia al resultado. Acepta solo dos resultados posibles: True (= 1) o False (= 0).
- none/null = valor nulo, indica que una variable no tiene valor y sirve para no utilizar el número cero para indicar esto.

Con la función built in ```type()``` podemos ver el tipo de dato del valor almacenado

````python
a = 2
b = "3"
c = 2.5

print(f"a = {a}", type(a))
print(f"b = {b}", type(b))
print(f"c = {c}", type(c)) 
````

**Conversiones de tipos de datos**

Se utilizan las siguientes funciones built in:
- int() convierte float o string a número entero
- float() convierte integer o string a número fraccionario
- str() convierte integer o float a cadena de caracter

````python
a_str = str(a)
a_float = float(a)
b_int = int(b)
b_float = float(b)
c_str = str(c)
c_int = int(c)

print(f"Integer to String: \na_str = {a_str}", type(a_str))
print(f"\nInteger to Float \na_float = {a_float}", type(a_float))
print(f"\nString to Integer \nb_int = {b_int}", type(b_int))
print(f"\nString to float \nb_float = {b_float}", type(b_float))
print(f"\nFloat to String \nc_str = {c_str}", type(c_str))
print(f"\nFloat to Integer \nc_int = {c_int}", type(c_int))
````

### Funciones

Son porciones de código que pueden contener desde una a varias lineas y desde ninguna a muchas variables, **identificadas** con un nombre represntativo de la tarea que esta ejecuta y que se pueden volver a reutilizar en el script.

Se crean con la palabra reservada ```def```.

Las **variables** que se pasan como **parámetros** a la función no deben tener un nombre ya utilizado, no deben definirse previamente en el script. Son nombres genéricos para trabajar en el **ámbito** interno de la función y los nombres de las variables se asignan al momento de invocar la función.

El resultado final, obtenido al procesar la función puede mostrarse por pantalla a través de la función **print** o almacenarse en otra variable, para un procesamiento posterior, a través del comando ```return```. Es posible almacenar más de un resultado pasando luego del comando **return** los nombres de sus variables separados por **comas**.

Funciones built in:
- input() = función de entrada de valores simples, siempre en tipo str, cuando el programa llega a una función input() el mismo se detiene y no reanudará su ejecuciónhasta que el usuario ingrese el valor solicitado.
- print() = función de salida simple que se emplea pasando como parámetro la variable cuyo valor se quiere mostrar en pantalla o el contenido mismo.


````python
# Función con print()
def to_lower_print (frase):
    minus = frase.lower()
    print(f"El texto convertido es: {minus}")

texto = "WITH MAYONESA PLEASE"

convertido = to_lower_print(texto)
````

## Lista, Tupla y Diccionario

### Lista

##### Definición

Una lista es un conjunto de elementos que pueden pertenecer al mismo tipo de dato, o ser una mezcla de ellos, y admite la repetición de los elementos. Se pueden almacenar: caracteres, letras, números enteros, númmeros decimales e incluso objetos como diccionarios e incluso listas construyendo asi una lista de listas.

Se definen con ```corchetes []``` y sus elementos se separan con ```comas```, donde el primer elemento tiene valor de posición **cero**.

El índice acepta tanto valores positivos como negativos, uno contará desde el elemento cero hacia la derecha y el otro hacia la izquierda.

````python
lista_numeros = [1, 2, 3, 4, 5]
lista_letras = ['a', 'b', 'c', 'd', 'e']
lista_palabras = ['arbol', 'casa', 'perro','auto']
lista_combinada = [1, 'a', 'arbol', 2, 'b', 'casa']


print("lista_numeros:",lista_numeros)
print("lista_letras:",lista_letras)
print("lista_palabras:",lista_palabras)
print("lista_combinada:",lista_combinada)
````

````python
# Índices
print(lista_palabras[2])
print(lista_palabras[-1])
````

````python
# Una lista es un tipo de dato en si
print(type(lista_combinada))

# Podemos ver como una lista mantiene el tipo de dato almacenado sin realizarle modificaciones
print(type(lista_combinada[2]))
print(type(lista_combinada[3]))
````

````python
lista_de_listas = [lista_numeros, lista_letras, lista_palabras]
print("lista_de_listas:",lista_de_listas)
````

````python
print(lista_de_listas[1])
````

````python
# Tambien se pueden crear listas vacías para su posterior llenado
lista_vacia = []
print(lista_vacia)

lista_vacia.append(2)
print(lista_vacia)
````

Operaciones ariméticas

- Suma: al sumar dos listas, se crea una lista resultante con la concatenación de las listas operadas una detrás de la otra
- Multiplicación: sólo admite escalares y el producto es una concatenación de la lista **x** cantidad de veces de forma consecutiva

````python
# Suma
a = [1, 2, 3]
b = ['a', 'b', 'c']
print("Suma:", a+b)

# Multiplicación
c= [1, 2, 3]
print("Multiplicación:" ,c*2)
````

### Diccionario

Es una estructura de datos que permite agrupar elementos no aislados, se define entre ```llaves {}``` y se declaran con el formato **clave:valor** separando cada parámetro con ```2 puntos :```(=key/value, llave/valor). El valor puede tratarse de cualquier tipo de dato y los pares key/value se separan con **comas**.

##### Definición

````python
usuario_1 = {
    "nombre":"Federico",
    "apellido":"Ford",
    "edad":"31",   
}

print(usuario_1) # La función print() visualmente no arroja diferencias pero devuelve el resultado en tipo cadena de texto
usuario_1
````

Para acceder a un elemento del diccionario no lo hacemos con índices, sino que lo hacemos a través de su clave. A pesar de que el diccionario se define entre llaves, para acceder a la clave lo hacemos con ```corchetes``` y entre ```comillas```. Es visualmente igual que un índice, pero técnicamente es una clave la que se llama para ver el valor asociado. No se hace por posición, sino por nombre.

````python
usuario_1["edad"]
````

Los diccionarios, al igual que las listas, pueden declararse vacíos para luego ir completándose a medida que el script se ejecuta

````python
dic_vacio = {}
print(dic_vacio)

dic_vacio["color"] = "Rojo"
print(dic_vacio)
````

##### Operaciones
Existen tres operaciones fundamentales que se pueden aplicar sobre un diccionario
- Agregar un elemento al diccionario: se invoca el diccionario y seguido entre corchetes se agrega la clave nueva que queremos introducir al mismo luego definimos el valor que esta tendrá guardado
- Modificar un elemento: se accede a la clave cuyo valor se desea modificar y se declara despues del operador **=** el nuevo valor a guardar
- Eliminar un elemento: en este caso primero se define la palabra reservada "del" luego el diccionario y la clave

````python
usuario = {
    "nombre":"federico",
    "apellido":"ford",
    "edad":"31", 
    "nacionalidad":"argentino"  
}

usuario
````

````python
# Agregar un elemento
usuario["genero"] = "masculino"
print(usuario)

# Modificar un elemento
usuario["nombre"] = "cesar federico"
print(usuario)

# Eliminar un elemento
del usuario["genero"]
print(usuario)
````

## Condicionales e Iteradores

### Valores Booleanos y Operadores de Comparación

Loa valores booleanos nos permiten modificar el flujo de programas y comenzar a dotar de "inteligencia" a nuestro código a través de la **lógica condicional**. A través de la lógica condicional se evalúan situaciones y de acuerdo al resultado el código sigue ejecutándose por una ruta u otra. A modo analógico podemos decir que el programa "hace preguntas" y "responde", luego de acuerdo a esta respuesta se determina el flujo del código.

Estas respuestas sólo aceptan dos valores posibles, el programa solamente puede identificar si algo es **True** (cierto, verdadero, si se cumple) o **False** (no cierto, falso, no se cumple).

##### Operadores de comparación

**Operadores con números (int y float)**
- Permiten todas las comparaciones: ==, !=, >, <, >=, <=.
- Combinaciones posibles: entre enteros, entre flotantes, o entre enteros y flotantes (Python hace conversión automática).

**Operadores con cadenas de caracteres (str)**
- Soportan: ==, !=, >, <, >=, <=.
- Solo se pueden comparar cadenas con otras cadenas, no con números ni listas
- Comparación de tipo ```lexicográfica```: Se comparan carácter por carácter usando el valor Unicode de cada letra. Si un carácter en la posición i es mayor que el correspondiente en la otra cadena, toda la cadena se considera mayor. Si son iguales, se pasa al siguiente carácter. Si se acaban los caracteres de una cadena antes que la otra, la más corta es menor.

**Operadores con diccionarios (dict)**
- Soportan: solo == y !=.
- Comparación: requiere coincidencia completa de claves y valores, no se comparan “parcialmente” ni de manera independiente.
- No se pueden usar: >, <, >=, <= con diccionarios (Python lanzará un error).

**Operadores con listas o tuplas**
- Soportan: ==, !=, >, <, >=, <=.
- Debe haber elementos comparables entre sí, números (int o float) con números y cadenas con cadenas. No soporta la comparacion de números con cadenas.
- Admite comparación de conjuntos con tipos mixtos, pero como Python compara elemento por elemento de manera paralela (es decir que cada comparación se hace entre los elementos en la misma posición) la primera diferencia entre pares de elementos determina el resultado de toda la comparación. Si son números la comparación es numérica, si son letras es lexicográfica y si un par de elementos no se puede comparar, Python lanza un error y no sigue.

### Condicionales If, Elif & Else

Existen tres tipos de condicionales:
- ```If```: Se ejecuta con resultado ```True```

*"Si la condición principal se cumple (es True) se ejecuta la acción siguiente"*-
- ```Else```: Se ejecuta con resultado ```False```

*"Si la condición principal no se cumple (es False) se ejecuta otra acción alternativa a la determinada como principal"*
- ```Elif```: Se ejecuta con resultado ```False``` seguido de un resultado ```True```

*"Si la condición princial no se cumple, pruebo estas condiciones alternativas y ejecuto las acciones siguientes."*

**If** solo se ejecuta cuando su resultado de comparación es **True**, entonces si yo lo que quiero es tener como condicion principal una diferencia tengo que escribirla con **!=** para que su resultado sea **True**. Si lo hubiese hecho con **==** al comparar elementos diferentes, el sentido es el mismo porque arrojaría **False** pero no cumpliría con el requisito del condicional **If** de ejecutarse solamente con resultados **True**.

Mientras que **Elif** permite incluir varios escenarios posibles y pueden declararse cuantas se deseen sin restricciones y es una práctica aceptada, con **If** sucede que también el código lo permite sin arrojar error pero esta sería una manera poco práctica de escribir código.

````python
a = 15

if a > 6:
    print("La variable es mayor que 6") #True
else:
    print("La variable es menor que 13") #False
````

````python
a = 5

if a > 6:
    print("La variable es mayor que 6") #False
else:
    print("La variable es menor que 13") #True
````

````python
a = 7

if a > 6:
    print("La variable es mayor que 6") #True
elif a < 13:
    print("La variable es menor que 13") #True (no se muestra porque cuando se cumple la primer condición if el condicional culmina)
````

````python
a = 4

if a > 6:
    print("La variable es mayor que 6") #False
elif a < 5:
    print("La variable es menor que 5") #True
````

````python
a = 5

if a > 6:
    print("La variable es mayor que 6") #False
elif a < 4:
    print("La variable es menor que 4") #False
else:
    print("La variable se encuentra entre 4 y 6") #True
````

**Expresiones booleanas compuestas**

La alternativa de Python para no anidar condiciones If (como en excel) fue introducir las keywords **and** y **or**.
- ```and```: tiene como resultado ```True``` cuando **todas** las condiciones se cumplen 
- ```or```: tiene como resultado ```True``` cuando al menos **una** de las condiciones se cumple 

````python
a = 5

if a > 4 and a < 6:
    print("La variable se encuentra entre 4 y 6") #True

else:
    print("La variable no se encuentra entre 4 y 6") #False
````

````python
# Ejemplo real: para hacer un servicio técnico al auto, el mismo debe cumplir al menos una de dos condiciones
# a) tener mas de 5 años sin servicio técnico o b) tener mas de 25000 km de recorrido
# supongamos un auto con 3 años sin servicio y 30000 km

# Condiciones para service
km = 25000
years = 5

km_auto = 30000
years_auto = 3

if years > years_auto or km < km_auto:
    print("Llevar el auto al servicio técnico")
else:
    print("No es necesario realizar un servicio técnico")
````

### Iteradores

**Ciclo While**
- Se abre con la palabra reservada **while**
- Se ejecuta siempre y cuando la condición se mantenga como **True**, cuando la condición evaluada inicialmente deja de cumplirse el ciclo deja de ejecutarse.
- Una variante es **While True** el cual no depende de ninguna condición, se declara manualmente que es verdadero y queda fijo. Su finalización se da con una sentencia **Break**.
- Cuándo usar cada uno:
    - while: cuando se puede definir claramente la condición de salida y depende de una sola variable o criterio simple.
    - while True: cuando la condición de salida es compleja, depende de múltiples factores o eventos externos, o querés controlarla manualmente dentro del bucle con break.

**Ciclo For**
- Se abre con la palabra reservada **for** (para mas detalle, ver el notebook "metodos_parametros_funciones")
- Se ejecuta una cantidad predefinida de veces que es igual a la de elementos iterables que se recorren.
- Es util para inspeccionar listas y diccionarios elemento a elemento para mostrarlos en pantalla o realizar modificaciones

````python
# While

iteraciones = 1

while iteraciones <= 5:
    print(f"Faltan {iteraciones} vueltas para finalizar el ciclo")
    iteraciones += 1

print("Ciclo finalizado")
````

````python
iteraciones = 6

while iteraciones <= 5: 
    print(f"Faltan {iteraciones} vueltas para finalizar el ciclo")
    iteraciones += 1

print("Ciclo finalizado")
````

````python
# For en lista

colores = ['rojo','azul', 'amarillo', 'verde']
for i in (colores):
    print(i)
````

````python
numeros = [1, 2, 3, 4, 5]

for i in (numeros):
    i += 10
    print(i)
````

````python
# For en diccionario

paciente = {
    "nombre":"juan",
    "apellido":"perez",
    "edad":"32",
    "genero":"masculino"
}

for i in (paciente): # Por defecto arroja las keys, es lo mismo que agregar el método
    print(i)
````

````python
paciente = {
    "nombre":"juan",
    "apellido":"perez",
    "edad":"32",
    "genero":"masculino"
}

for i in (paciente.values()):  # Podemos agregar distintos métodos de diccionarios para tener resultados diferentes
    print(i)
````

````python
paciente = {
    "nombre":"juan",
    "apellido":"perez",
    "edad":"32",
    "genero":"masculino"
}

for i in (paciente.items()): 
    print(i)
````

### Break & Continue

Son una alternativa para salir de un ciclo antes, es decirsin enfrentar la opción de finalización que tienen por default (while: cambio de true a false, for:fin de iterables). Existen dos tipos:

- **Break**: Cuando durante un ciclo llegamos a un break, el programa termina completamente el ciclo y sale del mismo para continuar ejecutando las lineas siguientes.

- **Continue**: En este caso, el código solo sale de la vuelta en la que se encuentra (sin ejecutar las indicaciones restantes a la misma) pero comienza a recorrer el elemento siguiente, no sale del ciclo completamente a diferencia de break, sino que lo continúa.

````python
# Sin break

for value in range(1,11):
    print(f"Iteración nro: {value}")
    if value == 3:
        print(f"Encontré el valor: {value}")
        print("Saliendo del ciclo")
        
````

````python
#Con break

for value in range(1,11):
    print(f"Iteración nro: {value}")
    if value == 3:
        print(f"Encontré el valor: {value}")
        print("Saliendo del ciclo")
        break
````

````python
# Sin continue

for value in range(1, 11):
    if value % 2 != 0:
        print(f"Salto el número impar: {value}")

    resultado = value ** 2  
    print(f"Procesando {value}, su cuadrado es {resultado}")
````

````python
# Con continue

for value in range(1, 11):
    if value % 2 != 0:
        continue

    resultado = value ** 2  
    print(f"Procesando {value}, su cuadrado es {resultado}")
````

````python
# Sin break

value = 1

while value <= 10:
    print(f"Iteración nro: {value}")

    if value == 3:
        print(f"Encontré el valor: {value}")
        print("Saliendo del ciclo")

    value += 1
````

````python
# Con break

value = 1

while value <= 10:
    print(f"Iteración nro: {value}")

    if value == 3:
        print(f"Encontré el valor: {value}")
        print("Saliendo del ciclo")
        break

    value += 1
````

````python
# Sin continue

valores = ["10", "n/a", "25", "", "7", "??", "40"]

i = 0

while i < len(valores):
    valor = valores[i]

    if not valor.isdigit():
        print(f"Valor inválido encontrado: {valor!r}. Lo salto.")
        i += 1

    numero = int(valor)
    doble = numero * 2
    print(f"Procesando {numero}: su doble es {doble}")

    i += 1
````

````python
# Con continue

valores = ["10", "n/a", "25", "", "7", "??", "40"]

i = 0

while i < len(valores):
    valor = valores[i]

    if not valor.isdigit():
        i += 1
        continue  

    numero = int(valor)
    doble = numero * 2
    print(f"Procesando {numero}: su doble es {doble}")

    i += 1
````
